```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

### Autocorrelation

Autocorrelation measures the serial correlation with a lagged time-series version of itself to capture any correlations within a time series and its past values.

```python
data = Path('GDP AUS_NZD.csv')
gdp_df = pd.read_csv(data)
gdp_df = gdp_df.set_index('Year')
gdp_df.head()
```

```python
gdp_df.plot()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

ax1.acorr(gdp_df['Australia'], maxlags=24, color='deeppink')
ax1.set_title("Australia's GDP")
ax1.set_xlabel('Lags', fontsize=12)

ax2.acorr(gdp_df['New Zealand '], maxlags=24, color='blue')
ax2.set_title("New Zealand's GDP")
ax2.set_xlabel('Lags', fontsize=12)

plt.suptitle('Autocorrelation')

plt.show()
```

An initial graphical analysis of the Autocorrelation values of Australia's and New Zealand's appears to show an absence of Autocorrelation.
